```python
df_line = df_base.copy()
df_line = df_line.set_index("code")
df_line.drop("K00-K93", inplace=True)
df_line.drop("N00-N99", inplace=True)
df_line.drop("U00-U99", inplace=True)
df_line.drop("E00-E90", inplace=True)
df_line.drop("G00-G99", inplace=True)
# df_line.drop("E00-E90", inplace=True)
df_line = df_line.reset_index()
```

```python
import matplotlib.lines as mlines
import numpy as np
import matplotlib.pyplot as plt

# draw line
def newline(p1, p2, color='black'):
    ax = plt.gca()
    l = mlines.Line2D([p1[0],p2[0]], [p1[1],p2[1]], color='green' if p1[1]-p2[1] > 0 else 'red', marker='o', markersize=6)
    ax.add_line(l)
    return l

fig, ax = plt.subplots(1,1,figsize=(6,14), dpi= 80)
fig.tight_layout()

ymin = 0
ymax = 32000
# Vertical Lines
ax.vlines(x=1, ymin=ymin, ymax=ymax, color='black', alpha=0.7, linewidth=1, linestyles='dotted')
ax.vlines(x=3, ymin=ymin, ymax=ymax, color='black', alpha=0.7, linewidth=1, linestyles='dotted')

# Points
ax.scatter(y=df_line['2017'], x=np.repeat(1, df_line.shape[0]), s=10, color='black', alpha=0.7)
ax.scatter(y=df_line['2020'], x=np.repeat(3, df_line.shape[0]), s=10, color='black', alpha=0.7)

# Line Segmentsand Annotation
for p1, p2, dc, c in zip(df_line['2017'], df_line['2020'], df_line.death_cause, df_line.code):
    newline([1,p1], [3,p2])
    ax.text(1-0.05, p1, str(round(p1)), horizontalalignment='right', verticalalignment='center', fontdict={'size':14})
    ax.text(3+0.05, p2, dc + ', ' + str(round(p2)), horizontalalignment='left', verticalalignment='center', fontdict={'size':14})

# 'Before' and 'After' Annotations
ax.text(1-0.05, ymax, 'BEFORE', horizontalalignment='right', verticalalignment='center', fontdict={'size':18, 'weight':700})
ax.text(3+0.05, ymax, 'AFTER', horizontalalignment='left', verticalalignment='center', fontdict={'size':18, 'weight':700})

# Decoration
ax.set_title("Cause of death : difference between 2017 and 2020", fontdict={'size':22})
ax.set_xticks([1,3])
ax.set_xticklabels(["2017", "2020"])

# Lighten borders
plt.gca().spines["top"].set_alpha(.0)
plt.gca().spines["bottom"].set_alpha(.0)
plt.gca().spines["right"].set_alpha(.0)
plt.gca().spines["left"].set_alpha(.0)
plt.show()
```
